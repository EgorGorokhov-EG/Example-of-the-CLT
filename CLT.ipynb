# An Example of the Central Limit Theorem
In this notebook I'll look at how the CLT works on example of Gamma Distribution

```python
import numpy as np
import scipy as sp
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
beta = 1.99

# Generating values for gamma distribution
x = np.linspace(stats.gamma.ppf(0.01, beta), stats.gamma.ppf(0.99, beta), 1000)

# Visualizing results: density of distribution and it's pdf 
plt.hist(stats.gamma.rvs(beta, size=1000), density=True, bins=20)
plt.plot(x, stats.gamma.pdf(x, beta), lw=2)
plt.xlabel('X')
plt.ylabel('f(X)')
plt.show()
```

Formulas for mathematical expectation and variance. beta = 1.99, alpha = 1

$E(X) = \frac{\beta}{\alpha}$ = 1.99

$D(X) = \frac{\beta}{\alpha^2}$ = 1.99

Distribution: https://ru.wikipedia.org/wiki/%D0%93%D0%B0%D0%BC%D0%BC%D0%B0-%D1%80%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5

```python
#  Calculating means for 1000 samples with sizes 3, 10, 100

means_3 = sorted([np.mean(stats.gamma.rvs(beta, size=3)) for i in np.arange(1000)])
means_10 = sorted([np.mean(stats.gamma.rvs(beta, size=10)) for i in np.arange(1000)])
means_100 = sorted([np.mean(stats.gamma.rvs(beta, size=100)) for i in np.arange(1000)])

# Distribution density for these means and pdf of Normal distribution with E(X) and D(X) we've got above

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(20, 7))

ax1.hist(means_3, density=True, bins=20)
ax1.plot(means_3, stats.norm.pdf(means_3, loc=1.99, scale=(1.99/3)**0.5))
ax1.set_title('n=3')
ax1.set_xlabel('X')
ax1.set_ylabel('f(X)')

ax2.hist(means_10, density=True, bins=20)
ax2.plot(means_10, stats.norm.pdf(means_10, loc=1.99, scale=(1.99/10)**0.5))
ax2.set_title('n=10')
ax2.set_xlabel('X')
ax2.set_ylabel('f(X)')

ax3.hist(means_100, density=True, bins=20)
ax3.plot(means_100, stats.norm.pdf(means_100, loc=1.99, scale=(1.99/100)**0.5))
ax3.set_title('n=100')
ax3.set_xlabel('X')
ax3.set_ylabel('f(X)')
```

As you can see, the CLT works pretty well with gamma distribution. And as expected, the bigger size of a samples the better approximation.
